   print(s)

            z      wayanad
            dtype: object
we can access an item by referring to the label.
EG:print(s["y"])
   #output:- ernakulam
key/value object as series:-
we can also use key/value object like a dictionary when creating a series.
EG:d={"a":1,"b":2,"c":3}
   sd=pd.Series(d)
   sd
   #output:-  0
           a	1
           b	2
           c	3
           dtype: int64
  sd=pd.series(d,index=["a","b"])
  print(sd)
  #output:-a  1

# combine()

```python
import pandas as pd
df1=pd.DataFrame([[1,2],[3,4]])
df2=pd.DataFrame([[5,6],[7,8]])

def myfunc(a,b):
  if (a.sum() > b.sum()):
    return a
  else:
    return b
print(df1.combine(df2,myfunc))
```

```python
df1=pd.DataFrame([["Emil","Tobias","Linus"],[16,14,10]])
df2=pd.DataFrame([["Emil"],[17]])
df1.update(df2)
df1
```

# pop()

```python
data={
    "Name":["Sally","Mary","John"],
    "Age":[50,40,30],
    "Qualified":[True,False,False]
}
df=pd.DataFrame(data)

#Remove column
df.pop("Age")
df
```

#ffill()

```python
data={
    "age":[20,23,None,25],
    "name":[None,"Raju","Ram","Nahi"]
}
df=pd.DataFrame(data)
df.ffill()
```

# isnull().sum()

```python
data={
    "age":[20,23,None,25],
    "name":[None,"Raju","Ram","Nahi"]
}
df=pd.DataFrame(data)
print(df.isnull().sum())
```

# Groupby

```python
import pandas as pd
df=pd.DataFrame({
    "category":["A","B","A","B","A"],
    "value":[10,20,30,40,50]
})
df.groupby("category")["value"].sum()
```

#HISTOGRAM

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("/content/data.csv")
df.plot()
plt.show()
```

```python
plt.plot(df["Calories"])
```

```python
plt.figure(figsize=(8,5))
plt.plot(df["Calories"],color="red",marker=".")
plt.title(" Line Plot of Calories")
plt.xlabel("Index")
plt.ylabel("Calories")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df.index,df["Calories"],color="red",marker="x")
plt.title(" Scatter Plot of Calories")
plt.xlabel("Index")
plt.ylabel("Calories")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("/content/data.csv")
df.plot(kind="scatter",x='Duration',y="Calories")
plt.show()
```

```python
df["Duration"].plot(kind='hist')
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Calories"],bins=20,color="purple",edgecolor="black")
plt.title("Histogram of Calories")
plt.xlabel("Calories")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
